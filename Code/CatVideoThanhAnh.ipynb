```python
import cv2
import os

# Đường dẫn video
video_path = "Data4.mp4"

# Tạo thư mục lưu ảnh
output_folder = "frames6"
os.makedirs(output_folder, exist_ok=True)

# Mở video
cap = cv2.VideoCapture(video_path)

# Kiểm tra nếu video không mở được
if not cap.isOpened():
    print("Không thể mở video. Kiểm tra đường dẫn!")
    exit()

# Lấy FPS thực tế của video
frame_rate = int(cap.get(cv2.CAP_PROP_FPS))
print(f"FPS của video: {frame_rate}")

# Số khung hình cần trích xuất mỗi giây
desired_fps = 1  
frame_skip = max(1, frame_rate // desired_fps)  # Tránh lỗi chia cho 0
print(f"Trích xuất 1 khung hình mỗi {frame_skip} frame")

frame_count = 0
saved_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Thoát khi hết video

    if frame_count % frame_skip == 0:
        # Xoay ảnh 90 độ sang phải (theo chiều kim đồng hồ)
        # frame = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)

        frame_path = os.path.join(output_folder, f"frame_{saved_count:04d}.jpg")
        cv2.imwrite(frame_path, frame)
        print(f"Đã lưu: {frame_path}")  # Kiểm tra xem ảnh có được lưu không
        saved_count += 1

    frame_count += 1

cap.release()
print(f"Đã trích xuất {saved_count} khung hình và lưu trong thư mục '{output_folder}'")
```
